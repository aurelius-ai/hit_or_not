### Step 3:  I use stFeatureExtraction() to extract the short term feature sequences of each song

```python
from pyAudioAnalysis import audioBasicIO
from pyAudioAnalysis import audioFeatureExtraction
import matplotlib.pyplot as plt
```

##### stFeatureExtraction() returns a numpy matrix of 34 rows and N columns
(N is the number of short-term frames that fit into the input audio recording)
#### This plots the feature sequences of the zero crossing rate and the signal energy. 

##### The frame size is 100 msecs (1/10 of a second) and the frame step is 500 msecs (1/2 a second). To read the audio samples, I call function readAudioFile() from the audioBasicIO.py file.

### Rihanna's 2012 Hit: Diamonds

```python
[Fs, song] = audioBasicIO.readAudioFile(song);
F, f_names = audioFeatureExtraction.stFeatureExtraction(song, Fs, 0.100*Fs, 0.500*Fs);
plt.subplot(2,1,1); plt.plot(F[0,:]); plt.xlabel('Frame no'); plt.ylabel(f_names[0]); 
plt.subplot(2,1,2); plt.plot(F[1,:]); plt.xlabel('Frame no'); plt.ylabel(f_names[1]); plt.show()
```
